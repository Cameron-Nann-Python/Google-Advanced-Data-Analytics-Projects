### Boxplots

Perform a check for outliers on relevant columns such as trip distance and trip duration. Remember, some of the best ways to identify the presence of outliers in data are box plots and histograms. 

**Note:** Remember to convert your date columns to datetime in order to derive total trip duration.  

```python
# Convert data columns to datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

**trip distance**

```python
# Create box plot of trip_distance
sns.boxplot(x=df['trip_distance'], data=df)
plt.title('Trip Distance');
```

```python
# Create histogram of trip_distance
bins = np.arange(0,31)
sns.histplot(df['trip_distance'], bins=bins)
plt.title('Trip Distance');
```

**total amount**

```python
# Create box plot of total_amount
from matplotlib.ticker import FuncFormatter

def dollar(x, pos):
    return f'${x:,.2f}'  # No decimals, comma separator

sns.boxplot(x=df['total_amount'])
plt.title('Total Amount')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar));
```

```python
# Create histogram of total_amount
bins = np.arange(-11, 101, 5)
sns.histplot(df['total_amount'], bins=bins)
plt.title('Total Amount')
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar));
```

**tip amount**

```python
# Create box plot of tip_amount
sns.boxplot(x=df['tip_amount'])
plt.title('Tip Amount')
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar))
plt.xticks(rotation=45);
```

```python
# Create histogram of tip_amount
bins = np.arange(0, 21)
sns.histplot(df['tip_amount'], bins=bins)
plt.title('Tip Amount')
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar))
plt.show()
```

**tip_amount by vendor**

```python
# Create histogram of tip_amount by vendor
bins = np.arange(0, 21)
sns.histplot(x=df['tip_amount'], 
             hue=df['VendorID'], 
             multiple='stack',
             bins=bins)
plt.title('Tip Amount by Vendor')
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar))
plt.show()
```

Next, zoom in on the upper end of the range of tips to check whether vendor one gets noticeably more of the most generous tips.

```python
# Create histogram of tip_amount by vendor for tips > $10
bins = np.arange(10, 21)
df_filtered = df[df['tip_amount'] > 10]
sns.histplot(x=df_filtered['tip_amount'], 
             hue=df_filtered['VendorID'],
             bins=bins,
             multiple='stack',
             data=df_filtered)
plt.title('Tip amount over 10 by Vendor')
plt.gca().xaxis.set_major_formatter(FuncFormatter(dollar))
plt.show()
```

**Mean tips by passenger count**

Examine the unique values in the `passenger_count` column.

```python
df['passenger_count'].sort_values().unique()
```

```python
# Calculate mean tips by passenger_count
round(df.groupby(['passenger_count'])['tip_amount'].agg(['mean']), 2)
```

```python
# Create bar plot for mean tips by passenger count

mean_tip_by_passenger = round(df.groupby(['passenger_count'])['tip_amount'].agg(['mean']), 2).reset_index()
plt.figure(figsize=(10,6))
ax = sns.barplot(x='passenger_count', 
                 y='mean',
                 palette='Greens_d',
                 data=mean_tip_by_passenger)
plt.gca().yaxis.set_major_formatter(FuncFormatter(dollar))
plt.title('Average Tip by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Average Tip (USD)')
ax.axline(xy1=(0,mean_tip_by_passenger['mean'].mean()), 
          slope=0, ls='--', 
          color='red', 
          label='global mean')
ax.legend()
plt.show()
```

**Create month and day columns**

```python
# Create a month column
df['month'] = df['tpep_dropoff_datetime'].dt.month_name()

# Create a day column
df['day'] = df['tpep_dropoff_datetime'].dt.day_name()
```

**Plot total ride count by month**

Begin by calculating total ride count by month.

```python
# Get total number of rides for each month
rides_per_month = df['month'].value_counts()
```

Reorder the results to put the months in calendar order.

```python
# Reorder the monthly ride list so months go in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
rides_per_month = rides_per_month.reindex(index=months)
rides_per_month
```

```python
# Show the index
rides_per_month.index
```

```python
# Create a bar plot of total rides per month
sns.barplot(x=rides_per_month.index, 
            y=rides_per_month)
plt.title('Total Rides per Month')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Number of Rides')
plt.show()
```

**Plot total ride count by day**

Repeat the above process, but now calculate the total rides by day of the week.

```python
# Repeat the above process, this time for rides by day
rides_per_weekday = df['day'].value_counts()
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
       'Friday', 'Saturday', 'Sunday']
rides_per_weekday = rides_per_weekday.reindex(index=days)
rides_per_weekday
```

```python
# Create bar plot for ride count by day
sns.barplot(x = rides_per_weekday.index, 
            y = rides_per_weekday)
plt.title('Total Rides Per Weekday')
plt.xlabel('Weekday')
plt.xticks(rotation=45)
plt.ylabel('Number of Rides')
plt.show()
```

**Plot total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
# Repeat the process, this time for total revenue by day
revenue_per_day = df.groupby(['day'])['total_amount'].sum().reindex(index=days)
revenue_per_day
```

```python
# Create bar plot of total revenue by day
sns.barplot(x=revenue_per_day.index, 
            y=revenue_per_day)
plt.title('Total Revenue per Day')
plt.xlabel('Weekday')
plt.xticks(rotation=45)
plt.ylabel('Total Revenue (USD)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(dollar))
plt.show()
```

**Plot total revenue by month**

```python
# Repeat the process, this time for total revenue by month
revenue_by_month = df.groupby('month')['total_amount'].sum().reindex(index=months)
revenue_by_month
```

```python
# Create a bar plot of total revenue by month
sns.barplot(x=revenue_by_month.index, 
            y=revenue_by_month)
plt.title('Revenue by Month')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Total Revenue (USD)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(dollar))
plt.show()
```

**Plot mean trip distance by drop-off location**

```python
# Get number of unique drop-off location IDs
df['DOLocationID'].nunique()
```

```python
# Calculate the mean trip distance for each drop-off location
mean_trip_by_do_location = (df.groupby(['DOLocationID'])['trip_distance']
                                    .agg(['mean'])
                                    .sort_values(by='mean'))

# Sort the results in ascending order by mean trip distance
mean_trip_by_do_location
```

```python
bins = [0, 5, 10, 15, 20, 25]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25']
mean_trip_by_do_location['distance_bin'] =pd.cut(mean_trip_by_do_location['mean'],
                                                bins=bins,
                                                labels=labels)
binned_trips = mean_trip_by_do_location.groupby('distance_bin')['mean'].sum().reset_index()
binned_trips
```

```python
# Create a bar plot of mean trip distances by drop-off location in ascending order by distance
plt.figure(figsize=(10,10))
plt.subplot(211)
ax = sns.barplot(x=mean_trip_by_do_location.index,
                 y=mean_trip_by_do_location['mean'],
                 order=mean_trip_by_do_location.index)
ax.set_xticklabels([])
ax.set_xticks([])
plt.title('Average Trip Distance by DO Location')
plt.xlabel('DO Location')
plt.xticks(rotation=90)
plt.ylabel('Average Trip Distance (miles)')

# create a plot that makes sense
plt.subplot(212)
sns.barplot(x='distance_bin',
            y='mean',
            data=binned_trips)
plt.title('Average Trip Distance by Distance Bin')
plt.xlabel('Average Trip Distance (miles)')
plt.ylabel('Number of Trips')

plt.tight_layout()
plt.show()
```

**Histogram of rides by drop-off location**

First, check to whether the drop-off locations IDs are consecutively numbered. For instance, does it go 1, 2, 3, 4..., or are some numbers missing (e.g., 1, 3, 4...). If numbers aren't all consecutive, the histogram will look like some locations have very few or no rides when in reality there's no bar because there's no location. 

```python
# Check if all drop-off locations are consecutively numbered
totalIDS = np.arange(1,265)
missing_values = [number for number in list(totalIDS) if number not in list(df['DOLocationID'].values)]
print(df['DOLocationID'].sort_values())
print(missing_values)
```

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python

# DOLocationID column is numeric, so sort in ascending order, Convert to string
DO_sorted_string = df['DOLocationID'].dropna().sort_values().astype('str').reset_index()


# Plot

import plotly.express as px

fig = px.histogram(DO_sorted_string, x='DOLocationID')
fig.update_layout(xaxis_tickangle=-90, title="Interactive Histogram of DOLocationIDs")
fig.show()
```
